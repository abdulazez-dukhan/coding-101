## Mental Model

Streamlit lets you write a Python script that behaves like a web app.

A normal Python script runs in the terminal and prints output.

A Streamlit script:

1. Runs top to bottom
2. Renders UI elements in the browser
3. Waits for user interaction
4. Reruns the script when the user changes something

### Big-picture mental model

```text
Python script
    ↓
Streamlit commands
    ↓
Interactive browser UI
    ↓
User changes input
    ↓
Script reruns
    ↓
Updated UI appears
```

### Streamlit is not a full frontend framework

Streamlit is not the same as React, Next.js, or Angular.

Instead, it is best understood as:

> Python-first UI for data apps and AI prototypes.

### How this concept fits production systems

In production AI systems, Streamlit usually appears as:

- an internal demo app
- a model testing interface
- a RAG playground
- a data-quality review tool
- an evaluation dashboard
- a lightweight admin interface

For customer-facing, highly custom interfaces, teams often use React or Next.js.

For internal AI engineering tools, Streamlit can be extremely efficient.

## Background Explanation

### Core concepts

#### 1. App script

A Streamlit app is usually a `.py` file.

Example:

```python
import streamlit as st

st.title("My App")
st.write("Hello from Streamlit")
```

You run it from the terminal:

```bash
streamlit run app.py
```

#### 2. Widgets

Widgets collect user input.

Examples:

```python
st.text_input("Question")
st.slider("Top K", 1, 10, 3)
st.button("Search")
st.selectbox("Model", ["small", "large"])
```

#### 3. Rerun model

Streamlit reruns the script from top to bottom after user interactions.

If you write:

```python
count = 0
if st.button("Add"):
    count += 1
st.write(count)
```

The count will reset on rerun.

To persist values, use `st.session_state`.

#### 4. Session State

`st.session_state` stores values across reruns for the current user session.

#### 5. Caching

Caching avoids repeating expensive work.

Use:

- `st.cache_data` for data outputs such as DataFrames, lists, dicts, search results
- `st.cache_resource` for global resources such as ML models, database clients, or vector DB clients

#### 6. Layout

Streamlit provides layout tools:

```python
st.sidebar
st.columns
st.tabs
st.expander
st.container
```

## Solution 2

````python
solution_2_code = """
import streamlit as st
import pandas as pd

st.set_page_config(page_title="Retrieval Review Dashboard", layout="wide")

@st.cache_data
def load_chunks():
    return pd.DataFrame(
        [
            {
                "chunk_id": "chunk-001",
                "source": "rag_notes.md",
                "text": "RAG retrieves relevant chunks before generation.",
                "score": 0.91,
            },
            {
                "chunk_id": "chunk-002",
                "source": "qdrant_notes.md",
                "text": "Vector databases store embeddings and metadata payloads.",
                "score": 0.88,
            },
            {
                "chunk_id": "chunk-003",
                "source": "streamlit_notes.md",
                "text": "Streamlit helps build fast internal AI demos.",
                "score": 0.84,
            },
            {
                "chunk_id": "chunk-004",
                "source": "rag_notes.md",
                "text": "Retrieved context should be shown with citations.",
                "score": 0.82,
            },
        ]
    )

if "history" not in st.session_state:
    st.session_state["history"] = []

df = load_chunks()

st.title("Retrieval Review Dashboard")
st.markdown(
    """
    This dashboard simulates a small tool for reviewing retrieved chunks.
    In a real RAG system, these rows would come from a vector database like Qdrant.
    """
)

source_options = ["All"] + sorted(df["source"].unique().tolist())
selected_source = st.sidebar.selectbox("Filter by source", source_options)

if selected_source == "All":
    filtered_df = df
else:
    filtered_df = df[df["source"] == selected_source]

st.subheader("Retrieved chunks")
st.dataframe(filtered_df, use_container_width=True)

question = st.text_input("Question to test", placeholder="Ask a question about RAG or Streamlit")

if st.button("Save question"):
    if question.strip():
        st.session_state["history"].append(question)
        st.success("Question saved to session history.")
    else:
        st.warning("Please enter a question first.")

st.subheader("Session question history")

if st.session_state["history"]:
    for i, item in enumerate(st.session_state["history"], start=1):
        st.write(f"{i}. {item}")
else:
    st.info("No questions saved yet.")
"""

solution_2_path = APP_DIR / "solution_2_retrieval_review_dashboard.py"
solution_2_path.write_text(solution_2_code, encoding="utf-8")

print(f"Created: {solution_2_path}")
print("Run:")
print(f"streamlit run {solution_2_path}")
````

## Common Mistakes

### Mistake 1: Running Streamlit apps like normal Python scripts

Wrong:

```bash
python app.py
```

Better:

```bash
streamlit run app.py
```

`python app.py` may execute the script, but it will not launch the proper Streamlit app experience.

### Mistake 2: Forgetting that Streamlit reruns the script

This does not persist correctly:

```python
count = 0

if st.button("Add"):
    count += 1

st.write(count)
```

Use `st.session_state` instead:

```python
if "count" not in st.session_state:
    st.session_state.count = 0

if st.button("Add"):
    st.session_state.count += 1

st.write(st.session_state.count)
```

### Mistake 3: Loading large models on every rerun

Bad:

```python
model = load_large_model()
```

Better:

```python
@st.cache_resource
def load_model():
    return load_large_model()

model = load_model()
```

### Mistake 4: Using `st.cache_data` for model objects

Use:

- `st.cache_data` for serializable data
- `st.cache_resource` for models, database clients, vector DB clients, and other shared resources

### Mistake 5: Building too much business logic inside the Streamlit file

Bad structure:

```text
app.py contains UI + retrieval + chunking + embedding + evaluation + database logic
```

Better structure:

```text
app.py
src/
  retriever.py
  embedder.py
  generator.py
  evaluation.py
```

Streamlit should mostly be the interface layer.

````python
common_mistakes_app = """
import streamlit as st

st.title("Session State Debugging Example")

if "count" not in st.session_state:
    st.session_state.count = 0

if st.button("Add"):
    st.session_state.count += 1

if st.button("Reset"):
    st.session_state.count = 0

st.write("Current count:", st.session_state.count)

st.markdown(
    """
    This example demonstrates why session state matters.
    Without session state, the counter would reset after each rerun.
    """
)
"""

common_mistakes_path = APP_DIR / "session_state_debugging_example.py"
common_mistakes_path.write_text(common_mistakes_app, encoding="utf-8")

print(f"Created: {common_mistakes_path}")
print("Run:")
print(f"streamlit run {common_mistakes_path}")
````

### Final Mini Exercise Solution

````python
final_solution_code = """
import streamlit as st
import pandas as pd

st.set_page_config(page_title="Mini RAG Playground", layout="wide")

@st.cache_data
def load_fake_chunks():
    return pd.DataFrame(
        [
            {
                "rank": 1,
                "score": 0.92,
                "source": "rag_intro.md",
                "text": "RAG retrieves relevant source chunks before generation.",
            },
            {
                "rank": 2,
                "score": 0.87,
                "source": "vector_db.md",
                "text": "Vector databases search by semantic similarity over embeddings.",
            },
            {
                "rank": 3,
                "score": 0.81,
                "source": "evaluation.md",
                "text": "RAG evaluation checks whether retrieved sources support the answer.",
            },
            {
                "rank": 4,
                "score": 0.76,
                "source": "streamlit.md",
                "text": "Streamlit is useful for building fast AI demos and review dashboards.",
            },
        ]
    )

st.title("Mini RAG Playground")
st.markdown(
    """
    This is a production-style mockup of a RAG demo interface.
    It does not call a real LLM or vector database, but the UI structure is realistic.
    """
)

top_k = st.sidebar.slider("Top K", min_value=1, max_value=4, value=3)
retrieval_mode = st.sidebar.selectbox("Retrieval mode", ["semantic", "hybrid", "keyword"])
show_debug = st.sidebar.checkbox("Show debug information", value=True)

question = st.text_area(
    "Ask a question",
    placeholder="How does RAG use a vector database?",
    height=120,
)

if st.button("Generate answer"):
    if not question.strip():
        st.warning("Please enter a question first.")
    else:
        chunks = load_fake_chunks().head(top_k)

        st.subheader("Generated answer")
        st.success(
            "A RAG system embeds the user question, retrieves the most relevant chunks "
            "from a vector database, and sends those chunks to the LLM so the answer is "
            "grounded in source material."
        )

        st.subheader("Retrieved chunks")
        for _, row in chunks.iterrows():
            with st.expander(f"Rank {row['rank']} | Score {row['score']} | Source: {row['source']}"):
                st.write(row["text"])

        if show_debug:
            st.subheader("Debug information")
            st.write("Retrieval mode:", retrieval_mode)
            st.dataframe(chunks, use_container_width=True)
else:
    st.info("Enter a question and click Generate answer.")
"""

final_solution_path = APP_DIR / "final_rag_playground_solution.py"
final_solution_path.write_text(final_solution_code, encoding="utf-8")

print(f"Created: {final_solution_path}")
print("Run:")
print(f"streamlit run {final_solution_path}")
````

- Streamlit reruns the script after user interaction.
- Use `st.session_state` to persist values across reruns.
- Use `st.cache_data` for data-loading and expensive data transformations.
- Use `st.cache_resource` for models, database clients, and global resources.
- Streamlit is excellent for internal AI tools, demos, dashboards, and prototypes.
- In production AI engineering, Streamlit is often the interface layer, not the full backend.
