# Mental Model

Think of `argparse` as a translator between:

## Terminal Commands

```bash
python app.py --phase chunk --book 8473
```

and Python objects:

```python
args.phase == "chunk"
args.book == 8473
```

---

`argparse` automatically:

1. Reads terminal input
2. Validates it
3. Converts it into Python variables
4. Stores them inside `args`


````python
import argparse
````


# Example 1 — Smallest Working Example

This is the smallest possible version of your pipeline idea.


````python

parser = argparse.ArgumentParser()

parser.add_argument("--phase")

args = parser.parse_args(args=["--phase", "chunk"])

print(args)
print("phase =", args.phase)
````


## What Happened?

This line:

```python
parser.add_argument("--phase")
```

creates a command-line argument.

This line:

```python
args = parser.parse_args(...)
```

creates the `args` object.

The argument name becomes an attribute automatically:

```python
args.phase
```



# Why We Use `args=[...]` in Jupyter

Normally `parse_args()` reads from the real terminal.

But inside notebooks we simulate terminal input using:

```python
parse_args(args=[...])
```

This acts exactly like typing commands in the terminal.


# Solution — Exercise 1


````python

parser = argparse.ArgumentParser()

parser.add_argument("--book", type=int)

args = parser.parse_args(args=["--book", "8473"])

print(args.book)
print(type(args.book))
````


Notice:

```python
type=int
```

automatically converts:

```python
"8473"
```

into:

```python
8473
```

This is extremely common in AI pipelines:

- epochs
- batch_size
- top_k
- port numbers
- model dimensions



# Example 2 — Using `choices=[]`

This is exactly what your ingestion pipeline uses.


````python

parser = argparse.ArgumentParser()

parser.add_argument(
    "--phase",
    choices=["extract", "structure", "chunk", "eval"]
)

args = parser.parse_args(args=["--phase", "chunk"])

print(args.phase)
````


## What `choices=[]` Does

It restricts allowed values.

Valid:

```bash
--phase chunk
```

Invalid:

```bash
--phase banana
```

This protects production pipelines from invalid execution states.


# Example 3 — Default Values

Your pipeline used:

```python
default=None
```

Let's see why.


````python

parser = argparse.ArgumentParser()

parser.add_argument(
    "--phase",
    default=None
)

args = parser.parse_args(args=[])

print(args.phase)
````


Since we passed:

```python
args=[]
```

the user did NOT provide `--phase`.

So argparse used the default:

```python

```

````python

parser = argparse.ArgumentParser()

parser.add_argument(
    "--phase",
    choices=["extract", "structure", "chunk", "eval"],
    default=None
)

args = parser.parse_args(args=[])

phases = [args.phase] if args.phase else [
    "extract",
    "structure",
    "chunk",
    "eval"
]

print(phases)
````


# Example 4 — Boolean Flags

Your pipeline also used:

```python
action="store_true"
```

Let's understand it.


````python

parser = argparse.ArgumentParser()

parser.add_argument(
    "--force",
    action="store_true"
)

args = parser.parse_args(args=["--force"])

print(args.force)
````


## What Happened?

If the flag exists:

```bash
--force
```

then:

```python
args.force == True
```

Otherwise:

```python
False
```

This is commonly used for:

- overwrite
- debug
- verbose
- GPU mode
- training mode
