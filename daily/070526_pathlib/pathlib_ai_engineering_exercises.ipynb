This notebook teaches `pathlib`, Python's modern object-oriented way to work with files and folders.

  - Python documentation: https://docs.python.org/3/library/pathlib.html

### Why it matters in real AI engineering

AI projects usually have many moving parts:

- raw datasets
- processed datasets
- model checkpoints
- embedding indexes
- vector DB storage
- logs
- evaluation reports
- configuration files

Using fragile string paths like `"data/" + filename` can easily break across Windows, Linux, Docker, notebooks, or deployment servers. `pathlib` gives you safer and cleaner path handling.

## Mental Model

Think of `Path` as a **smart file/folder object**.

Instead of treating paths as plain strings, `pathlib` treats them as objects with useful methods.

```python
from pathlib import Path

project_root = Path("my_project")
data_dir = project_root / "data"
file_path = data_dir / "examples.jsonl"
```

The `/` operator joins path parts correctly for your operating system.

### How this concept fits production systems

In production AI systems, paths are often reused across modules:

- ingestion script writes chunks to `data/processed/chunks.jsonl`
- embedding script reads chunks and writes vectors to `indexes/`
- evaluation script writes metrics to `reports/eval_results.json`
- API loads models/configs from stable project paths

`pathlib` helps keep these paths predictable, readable, and portable.

## Background Explanation

### Core concepts

| Concept | Meaning |
|---|---|
| `Path` | Object representing a file or directory path |
| `/` operator | Joins path parts safely |
| `.exists()` | Checks whether a path exists |
| `.is_file()` | Checks whether path is a file |
| `.is_dir()` | Checks whether path is a directory |
| `.mkdir()` | Creates a directory |
| `.read_text()` | Reads text from a file |
| `.write_text()` | Writes text to a file |
| `.glob()` | Finds files matching a pattern |
| `.rglob()` | Recursively finds files matching a pattern |
| `.suffix` | File extension, such as `.jsonl` |
| `.stem` | Filename without extension |
| `.parent` | Parent folder |

### Real examples

```python
Path("data/raw/book.txt").suffix      # .txt
Path("data/raw/book.txt").stem        # book
Path("data/raw/book.txt").parent      # data/raw
```

In AI engineering, this helps you avoid hard-coded path strings scattered everywhere.

## Minimal Working Example

This example creates a tiny project-like folder, writes a text file, reads it back, and checks that it exists.

````python
from pathlib import Path

# Create a small workspace for this notebook
workspace = Path("pathlib_workspace")
data_dir = workspace / "data"

# Create folders safely
# parents=True creates missing parent folders
# exist_ok=True prevents an error if the folder already exists
data_dir.mkdir(parents=True, exist_ok=True)

# Create a file path
example_file = data_dir / "sample.txt"

# Write text
example_file.write_text("Pathlib is useful for AI engineering projects.", encoding="utf-8")

# Read text
content = example_file.read_text(encoding="utf-8")

print("File path:", example_file)
print("Exists:", example_file.exists())
print("Content:", content)
````

## Common Mistakes

### Mistake 1: Joining paths as strings

```python
bad_path = "data" + "raw" + "file.txt"
```

This produces:

```text
datarawfile.txt
```

Better:

```python
good_path = Path("data") / "raw" / "file.txt"
```

### Mistake 2: Forgetting `parents=True`

```python
Path("a/b/c").mkdir()
```

This fails if `a/b` does not already exist.

Better:

```python
Path("a/b/c").mkdir(parents=True, exist_ok=True)
```

### Mistake 3: Forgetting encoding

For English-only files, the default may work. For Arabic, multilingual datasets, or scraped text, always be explicit:

```python
text = path.read_text(encoding="utf-8")
```

### Mistake 4: Assuming current working directory

In notebooks, scripts, terminals, and APIs, the current working directory can change.

Useful debugging:

```python
Path.cwd()
```

````python
from pathlib import Path

print("Current working directory:")
print(Path.cwd())
````

### Suggested Solution for Final Mini Exercise

````python
from pathlib import Path

root = Path("production_rag_project")

PROJECT_PATHS = {
    "root": root,
    "raw_data": root / "data" / "raw",
    "processed_data": root / "data" / "processed",
    "chunks": root / "data" / "chunks",
    "models": root / "models",
    "reports": root / "reports",
    "eval_data": root / "eval" / "data" / "eval.jsonl",
    "vector_db": root / "storage" / "vector_db",
}

for name, path in PROJECT_PATHS.items():
    if path.suffix:  # likely a file path, such as .jsonl
        path.parent.mkdir(parents=True, exist_ok=True)
        path.touch(exist_ok=True)
    else:
        path.mkdir(parents=True, exist_ok=True)

for name, path in PROJECT_PATHS.items():
    path_type = "file" if path.is_file() else "dir" if path.is_dir() else "missing"
    print(f"{name:15} | {path_type:7} | exists={path.exists()} | {path}")
````

## Key Takeaways

- `pathlib.Path` is the modern Python way to work with files and folders.
- Use `/` to join paths instead of manual string concatenation.
- Use `.mkdir(parents=True, exist_ok=True)` for safe directory creation.
- Use `.read_text()` and `.write_text()` with `encoding="utf-8"`, especially for multilingual AI projects.
- Use `.glob()` and `.rglob()` to find files in ingestion pipelines.
- Use `.stem`, `.suffix`, and `.parent` to generate clean output filenames.
- In real AI engineering, good path management prevents many bugs in RAG, ML training, APIs, evaluation, and deployment.
